```python
#Cleaning Gender Column, Determining unique values and counts
print(f"Total number of unique values: {df_raw['Gender'].nunique()}")
print(f"Unique values: {df_raw['Gender'].unique()}")

#Counting number of each values
print("Number of each value:")
print(df_raw['Gender'].value_counts())
```

```python
#Variant Map

gender_map = {
    'Male' : 1,
    'Female' : 0,
    'MALE' : 1,
    'FEMALE' : 0,
    '1' : 1,
    '0' : 0
}

df_raw['Gender_cleaned'] = df_raw['Gender'].map(gender_map)

#Cleaning Verification
print("Cleaned Gender:")
print(df_raw['Gender_cleaned'].value_counts())

#Showcase Null Values if any
print(f"\nNaN value count: {df_raw['Gender_cleaned'].isnull().sum()}")
```

```python
#Removing Dirty Column
df_raw = df_raw.drop(columns=['Gender'])
df_raw = df_raw.rename(columns={'Gender_cleaned' : 'Gender'})

print("Dirty Column Removed")
print("Gender column after cleaning:")
print(df_raw['Gender'].value_counts())
print(f"Data type: {df_raw['Gender'].dtypes}")
df_raw.head()
```
